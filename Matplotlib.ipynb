```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
output_rate_of_tomato = [sum(map(ord, month)) % 100 for month in months]

output_rate_of_tomato
```

```python
plt.title('Tomato Rate From January to December')
plt.xlabel('Month of 2024', loc='center')
plt.ylabel('Tomato Production Rate In Each Month', loc='center')
plt.plot(*(months, output_rate_of_tomato))
```
