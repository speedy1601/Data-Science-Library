```python
from skimage import io, filters, color, data, transform
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
```

# Coverting a rgb/grayscale image to a Binary(0/1) image where values are 0(black) and 1(white) `using threshold`

```python
image = io.imread(r"C:\Users\user\Downloads\handwritten.jpg")
grayImage = color.rgb2gray(image)

threshold_otsu = filters.threshold_otsu(grayImage)
print("threshold_otsu =", threshold_otsu)
binaryImage = grayImage > threshold_otsu # A Boolean Comparison, [[True, False...]] which is a Binary(0/1) Image.

threshold_local = filters.threshold_local(grayImage, block_size=15, method='gaussian')
print("threshold_local =", threshold_local)
binaryImage_1 = grayImage > threshold_local

plt.figure(figsize=(10, 7)) # inches
gs = GridSpec(nrows=2, ncols=2, hspace=0.01)

plt.subplot(gs[0, :])
plt.title("Grayscale Image")
plt.imshow(grayImage, cmap='gray')
plt.axis('off')

plt.subplot(gs[1, 0])
plt.title("Binary(0/1) Image Otsu")
plt.imshow(binaryImage, cmap='gray')
plt.axis('off')

plt.subplot(gs[1, 1])
plt.title("Binary(0/1) Image Local")
plt.imshow(binaryImage_1, cmap='gray')
plt.axis('off')
```
