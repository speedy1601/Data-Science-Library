# Page segmentation modes:
0. Orientation and script detection (OSD) only.
1. Automatic page segmentation with OSD.
2. Automatic page segmentation, but no OSD, or OCR.
3. Fully automatic page segmentation, but no OSD. (Default)       ----- It merge `One or multiple linesapce` into `single linespace`
4. Assume a single column of text of variable sizes.
5. Assume a single uniform block of vertically aligned text.
6. Assume a single uniform block of text.                         ----- It ignores the `linesapce`
7. Treat the image as a single text line.
8. Treat the image as a single word.
9. Treat the image as a single word in a circle.
10. Treat the image as a single character.
11. Sparse text. Find as much text as possible in no particular order.
12. Sparse text with OSD.
13. Raw line. Treat the image as a single text line, bypassing hacks that are Tesseract-specific.

From the Above Code, "x1, y2, x2, y1 = map(int, line.split()[1:-1])", here after x1 its 'y2' not 'y1'. After pasting 'random1.jpg'
on 'paint' software, I saw that for example in `M 69 318 146 389 0`, 69 = x1 and 318 = y2.. x1 and x2 placed correctly but y1 and y2 are not placed correctly..

Now width is calculated from Left to Right in everywhere. In cv2 the height is counted from 'top to bottom' but here 'y1 and y2' is counted from 'bottom to top' which is done by `tesseract.image_to_boxes(image)`. So here tesseract calculate height in opposite way of cv2. But if we want to pass this 'y1, y2' to a cv2 function, calculated by 'tesseract.image_to_boxes()', then we have to convert 'y1, y2' in cv2 format i.e. 'height - y1, height - y2' (height = img.shape[0])

`Summary :`
            `x1, y2, x2, y1 = map(int, line.split()[1:-1])` # From `M 69 318 146 389 0`, [69, 318, 146, 389]
            `Top Left Coordinate, Bottom Right Coordinate = (x1, height - y1), (x2, height - y2)`

But IDK 'x1, y1, x2, y2 = map(int, line.split()[1:-1])' also works here for `cv2.rectangle`, not for `cv2.puttext`.

#                         Detect The WOrd and Put Bounding Box and Text on Detected Words using `pytesseract.image_to_data(image)`

````python
def doit(img) -> None:
    box: str = pytesseract.image_to_data(img, config=r'--psm 8 --oem 3') # for only digits  config=r"--psm 4 --oem 3 outputbase digits"
    # x, y, w, h = map(int, box.splitlines()[-1].split()[6:10]); i = 2
    # img = img[y-i:y+h+i, x-i:x+w+i] # cropping the main info

    box: str = pytesseract.image_to_data(img)
    print(pytesseract.image_to_data(img, config=r'--psm 8 --oem 3'), '\n')
    print(box, '\n')
    print(pytesseract.image_to_string(img), '\n')

    for lineNo, line in enumerate(box.splitlines()):
        line = line.split()
        if lineNo != 0 and len(line) == 12:
            x1, y1, w, h = map(int, line[6:10]) # explanations in below block
            cv2.rectangle(img, (x1, y1), (x1 + w, y1 + h), (0, 0, 0), 1)
            cv2.putText(img, line[-1], (x1, y1), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 1, cv2.LINE_AA)

    cv2.imwrite('Black and White Image.jpg', img)
    cv2.imshow('Detection', img)
    cv2.waitKey(0)

img = cv2.imread("images/random text.jpg")
grayImg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
adaptive_thrsehold = cv2.adaptiveThreshold(grayImg, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, blockSize=55, C=4)
threshold, black_and_white_img = cv2.threshold(grayImg, 80, 255, cv2.THRESH_BINARY_INV)

mainImg = adaptive_thrsehold
doit(mainImg)
````

Here in the above black, when we did `print(box)`, we saw levels from where `left, top` = `Top Left Corner of the Word` = `x1, y1`. This `y1` calculated from `Top to Bottom` which is how cv2 calculates height. In `tesseract.image_to_boxes` we saw `y1, y2` are calculated from `Bottom to Top` which is different from this (tesseract.image_to_data).

`width, height` = `width and height of the Bounding Box which we will put on that Word`
`x2, y2` = `Bottom Right corner of the Word` = `(x1 + width, y1 + height)`
